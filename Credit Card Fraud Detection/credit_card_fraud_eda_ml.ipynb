```python
#separating data and target

X = df.drop('Class', axis=1)
y = df['Class']
```

The dataset was divided into independent variables (features) and the target variable (Class), where Class indicates fraudulent or legitimate transactions.

```python
#Train-Test Split

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.3,
    random_state=42,
    stratify=y  #Ensures fraud ratio remains the same in train and test sets.
)
```

Stratified splitting was used to preserve the original class distribution in both training and testing datasets.

```python
#Feature Scaling

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

Feature scaling was applied to standardize numerical values and improve model performance.

```python
#model testing

from sklearn.linear_model import LogisticRegression

model = LogisticRegression(
    max_iter=1000,
    class_weight='balanced'
)

model.fit(X_train_scaled, y_train)
```

Class weighting was applied to address data imbalance and improve fraud detection performance.

```python
y_pred = model.predict(X_test_scaled)
```

```python
#Model Evaluation

#confusion matrix

from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred)
cm
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
```

Due to the highly imbalanced nature of the dataset, accuracy alone is not a reliable metric. Precision and recall were used to evaluate the model’s ability to correctly identify fraudulent transactions. The model achieved reasonable recall, indicating its effectiveness in detecting fraud cases.

```python
#visualizing confusion Matrix

plt.figure(figsize=(5,4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()
```
